```python
import pandas as pd
import json
from pandas.io.json import json_normalize

with open('insta.json', 'r') as f:
    data = json.load(f)
final_data = json_normalize(data['data'])
final_data
```

```python
tags_data = json_normalize(data=data['data'], record_path='tags')
tags_data
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
import nltk
nltk.download('vader_lexicon')
```

```python
sid = SentimentIntensityAnalyzer()
sentiments = []
df = pd.DataFrame()
```

```python
for row in tags_data.itertuples():
    print(row[1])
    ss = sid.polarity_scores(row[1])
    for k in ss:
        print('{0}: {1}, '.format(k, ss[k]), end='\n')
        sentiments.append([ss['neg'], ss['pos'],ss['neu'], ss['compound']])
df[['neg', 'pos', 'neu', 'compound']] = pd.DataFrame(sentiments)
```
